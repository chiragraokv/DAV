```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'mtcars.csv'
car_data = pd.read_csv(path)
car_data = car_data.dropna(subset=['mpg'])
car_data
```

```python
tab1 = pd.crosstab(car_data['cyl'],car_data['gear'])
print("table summarising number of cars (cylinders and gears)")
tab1
```

```python
sns.heatmap(tab1,annot=True,fmt='d')
plt.title("Crosstab heatmap")
plt.xlabel("Gears")
plt.ylabel("cylinders")
```

```python
tab2  = car_data[['model','mpg']].sort_values('mpg',ascending= False)
print("display all models from best to worst fuel efficiency")
tab2
```

```python
tab3 = car_data.groupby('cyl')['mpg'].mean()
print("Table displaying number of cylinders and average mpg")
tab3
```

```python
sns.barplot(x = 'cyl',y= 'mpg',data=car_data,ci = None)
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(car_data['mpg'], bins=10, kde=True)
plt.title("Distribution of MPG")
plt.show()
```

```python
car_data['mpg'] = pd.to_numeric(car_data['mpg'], errors='coerce')
car_data = car_data.dropna(subset=['mpg'])

plt.figure(figsize=(6,4))
sns.histplot(data=car_data, x='mpg', bins=10, kde=True)
plt.xlabel("Miles Per Gallon (mpg)")
plt.ylabel("Frequency")
plt.title("MPG Distribution")
plt.show()
```

```python
fig,axes = plt.subplots(1,2,figsize = (14,6))
#histogram of mpg
sns.histplot(car_data['mpg'],bins = 20)
```

```python
sns.pairplot(car_data[['mpg', 'hp', 'wt', 'qsec']])
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.violinplot(x='cyl', y='mpg', data=car_data)
plt.title("MPG Distribution by Cylinder Count")
plt.show()
```

```python
sns.regplot(x = 'wt',y = 'mpg',data = car_data,color= 'red')
plt.title("MPG vs Weights(Regression)")
```

```python
sns.swarmplot(x = 'cyl',y='mpg',data= car_data,color='green')
plt.title("Mpg by cylinder count")
plt.show()
```
