```python
#import the libraries such as numpy, pandas, matplotlib and seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Upload the dataset
df=pd.read_csv('mtcars.csv')
pd.set_option('display.max_columns', 100)  #display all columns
pd.set_option('display.max_rows',None)      #display all rows
```

```python
#View the dataset
df
```

```python
# 1. Extract the 'mpg' column from the dataframe & Plot the mpg values as a line plot
mpg = df['mpg']
mpg.plot()
# The default plot will be a line plot where the x-axis represents the index of the data points
# and the y-axis represents the 'mpg' values.
```

```python
# 2. Select only the 'cyl', 'wt', and 'mpg' columns from the dataframe & Plot the selected dataframe columns
df = df[['cyl', 'wt', 'mpg']]
df.plot()

# By default, Pandas will create a line plot for all numerical columns ('cyl', 'wt', 'mpg')
# - The x-axis represents the index of the dataframe.
# - The y-axis represents the values of 'cyl', 'wt', and 'mpg'.
# - Each column gets a separate line in the plot.
```

```python
#3. Bar Plot- the count of cars by the number of cylinders
cyl_counts = df['cyl'].value_counts()
cyl_counts.plot(kind='bar', color='green')
plt.title('Count of Cars by Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
#use ggplot for better visualization. ggplot2 is based on the Grammar of Graphics, which provides a structured approach to building plots.
# This makes it easy to add, remove, or modify components (e.g., layers, scales, themes) without breaking the code.
plt.style.use('ggplot')
```

```python
#4. Scatter plot of weight vs mpg
df.plot(
    kind="scatter",
    x="wt",
    y="mpg",
    figsize=(9,9),
    color="black",
);
```

```python
#5. Box plot of MPG for each cylinder category
plt.figure(figsize=(8, 6))
sns.boxplot(x='cyl', y='mpg', data=df)
plt.title('MPG by Cylinder Category')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
# 6. Histogram of MPG for all cars

x = df['mpg']

# Plotting the histogram
plt.hist(x)

# Adding title to the plot
plt.title("Distribution of MPG in Cars Dataset")

# Adding label on the y-axis (Frequency of each cylinder count)
plt.ylabel('Frequency')

# Adding label on the x-axis (Number of cylinders in cars)
plt.xlabel('MPG')

# Displaying the plot
plt.show()
```

```python
# 7. This pie chart shows the proportion of cars with 4, 6, or 8 cylinders.
# Group data by 'cyl' and calculate counts
cyl_counts = df['cyl'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    cyl_counts,
    labels=cyl_counts.index,
    autopct='%1.1f%%',  # Add percentage labels
    colors=['skyblue', 'lightgreen', 'orange'],
    startangle=140,  # Start angle of the pie chart
    explode=[0.1, 0, 0]  # Emphasize the first slice (4 cylinders)
)
plt.title('Proportion of Cars by Number of Cylinders')
plt.show()

```

```python
#8. Pie Chart for Transmission Types

# Reload the original dataframe to include the 'am' column
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mtcars(in).csv')


am_counts = df['am'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    am_counts,
    labels=['Automatic', 'Manual'],  # Custom labels
    autopct='%1.1f%%',
    colors=['gold', 'lightcoral'],
    startangle=90
)
plt.title('Proportion of Cars by Transmission Type')
plt.show()
```

```python
# 9. Nested Pie Chart which needs Data preparation
outer_labels = df['cyl'].value_counts().index
outer_sizes = df['cyl'].value_counts().values

inner_labels = df['gear'].value_counts().index
inner_sizes = df['gear'].value_counts().values

# Create nested pie chart
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(outer_sizes, labels=outer_labels, radius=1, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'orange'])
ax.pie(inner_sizes, labels=inner_labels, radius=0.7, autopct='%1.1f%%', colors=['pink', 'purple', 'cyan'])
ax.set(aspect="equal")
plt.title('Nested Pie Chart: Cylinders (Outer) and Gears (Inner)')
plt.show()
```
