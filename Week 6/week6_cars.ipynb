```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
car = pd.read_csv("mtcars.csv")
car
```

```python
car.info
```

```python
x_mpg =np.array(car['mpg'])
y_model = car['model']
```

```python
#1. Explore the distribution of fuel efficiency in the cars through a plot.
car["mpg"].plot(kind='hist', bins=20, title="Histogram of the distribution of MPG")
```

```python
#Best and worst fuel efficiency
best_mpg_car = car.loc[car["mpg"].idxmax(), ["model", "mpg"]]
worst_mpg_car = car.loc[car["mpg"].idxmin(), ["model", "mpg"]]
print("Best MPG Car:", best_mpg_car)
print("Worst MPG Car:", worst_mpg_car)
```

```python
#Categories of cars (cylinders)
car_categories = car["cyl"].value_counts()
plt.bar(car_categories.index, car_categories.values, color='red', edgecolor='black', alpha=0.7)
plt.xlabel("Number of Cylinders")
plt.ylabel("Count of Cars")
plt.title("Car Categories Based on Number of Cylinders")
plt.xticks(car_categories.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print(car_categories)
```

```python
#  Car with the worst horsepower
worst_hp_car = car.loc[car["hp"].idxmin(), ["model", "hp"]]
print("Car with Worst Horsepower:", worst_hp_car)
```

```python
#Five-number summary & box plot for displacement
five_number_summary = car["disp"].describe()[["min", "25%", "50%", "75%", "max"]]
plt.boxplot(car["disp"], vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.xlabel("Displacement (disp)")
plt.title("Box Plot of Displacement")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
print("Five-number summary for Displacement:\n", five_number_summary)
```

```python
car['am'].value_counts()
```

```python
#Heaviest car and its number of gears
heaviest_car = car.loc[car["wt"].idxmax(), ["model", "wt", "gear"]]
print("Heaviest Car:", heaviest_car)
```

```python
#Car with the best quarter-mile time (qsec)
best_qsec_car = car.loc[car["qsec"].idxmin(), ["model", "qsec"]]
print("Car with Best qsec:", best_qsec_car)
```

```python
#Average mpg for manual vs automatic cars
avg_mpg_transmission = car.groupby("am")["mpg"].mean()
avg_mpg_transmission.index = ["Automatic", "Manual"]
print("Average MPG for Transmission Types:\n", avg_mpg_transmission)
```

```python
#Side-by-side box plot of fuel efficiency for manual vs automatic cars
plt.boxplot([car[car["am"] == 0]["mpg"], car[car["am"] == 1]["mpg"]], labels=["Automatic", "Manual"])
plt.xlabel("Transmission Type")
plt.ylabel("Miles Per Gallon (mpg)")
plt.title("Fuel Efficiency Comparison: Manual vs Automatic Cars")
plt.show()
```

```python
#Relationship between car rate (drat) and mpg (scatter plot)
plt.scatter(car["drat"], car["mpg"], color="purple")
plt.xlabel("Rear Axle Ratio (drat)")
plt.ylabel("Miles Per Gallon (mpg)")
plt.title("Scatter Plot: Car Rate vs. MPG")
plt.show()
```

```python
# Categorize cars based on number of gears
gear_categories = car["gear"].value_counts()
plt.bar(gear_categories.index, gear_categories.values, color="orange")
plt.xlabel("Number of Gears")
plt.ylabel("Count of Cars")
plt.title("Car Categories Based on Number of Gears")
plt.xticks(gear_categories.index)
plt.show()
```

```python
# Relationship between fuel efficiency and number of gears
plt.bar(car["gear"], car["mpg"], color="red")
plt.xlabel("Number of Gears")
plt.ylabel("Miles Per Gallon (mpg)")
plt.title("Scatter Plot: Fuel Efficiency vs. Number of Gears")
plt.show()
```

```python
# Relationship between horsepower and number of cylinders
plt.bar(car["cyl"], car["hp"], color="blue")
plt.xlabel("Number of Cylinders")
plt.ylabel("Horsepower (hp)")
plt.title("Scatter Plot: Horsepower vs. Cylinders")
plt.show()
```

```python
#Relationship between displacement and gross horsepower
plt.scatter(car["disp"], car["hp"], color="green")
plt.xlabel("Displacement")
plt.ylabel("Horsepower (hp)")
plt.title("Scatter Plot: Displacement vs. Horsepower")
plt.show()
```

```python
#Best car for high speed and good fuel efficiency
best_speed_fuel_car = car.loc[(car["qsec"] == car["qsec"].min()) & (car["mpg"] == car["mpg"].max()), ["model", "qsec", "mpg"]]
print("Best Car for Speed & Fuel Efficiency:", best_speed_fuel_car)
```
