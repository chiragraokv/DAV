```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("deliveries.csv")
df.columns
```

```python
#1. Find and visualize the total runs scored by each team. 
tab = df.groupby("batting_team")["total_runs"].sum()
tab 
```

```python
sns.barplot(x = "batting_team",y = "total_runs",data = df)
plt.xticks(rotation = 90)
```

```python
#2. Identify and display the names of the top 10 highest-scoring batters with their total runs 
tab.sort_values(ascending = False).head(10)
```

```python
#. Find and visualize the top 10 bowlers with the most wickets.
bowler_tab = df.groupby("bowler")["is_wicket"].sum()
bowler_tab = bowler_tab.sort_values(ascending = False).head(10)
bowler_data = bowler_tab.reset_index()
bowler_data
```

```python
sns.barplot(x = "bowler",y = "is_wicket",data = bowler_data)
plt.xticks(rotation = 90)
```

```python
#4. Tabulate and visualize the total number of extras conceded by each team.
extras = df.groupby("batting_team")["extra_runs"].sum()
extras_data = extras.reset_index()
sns.barplot(x = "batting_team", y = "extra_runs", data = extras_data )
plt.xticks(rotation = 90)
```

```python
#5. Tabulate and visualize the batsmen (top 5) involved in the most run-outs. 
data_runout = df[df["dismissal_kind"] == "run out"]
data_runout = data_runout["batter"].value_counts()
data_runout = data_runout.head(10)
data_runout = data_runout.to_frame()
print(data_runout.columns)
```

```python
sns.barplot(x = "index", y = "run outs",data = data_runout)
```
