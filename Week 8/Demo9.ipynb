```python
data = pd.read_csv("mtcars.csv", index_col=0)
mpg = data["mpg"]
bins = [0,18,25,40]
labels = ["low", "medium","high"]
mpg_cat=pd.cut(mpg, bins = bins, labels = labels)
data['mpgCat']= mpg_cat
data
```

```python
tab3 = data.groupby('mpgCat')['wt'].mean()
print("Table displaying categories of MPG and their average weight")
print(tab3)
```

```python
df1 = data[['mpgCat','cyl', 'wt', 'mpg']]
plt.figure(figsize=(8, 6))
sns.boxplot(x='mpgCat', y='wt', data=df1)
plt.title('Weight by MPG Category')
plt.xlabel('MPG Category')
plt.ylabel('Weight')
plt.show()
```
